```python
# Problem Statement 1: Mobile Recharge Verification Program

# Function that checks whether the recharge request is valid
def validate_recharge(amount, selected_plan):
    valid_plans = [199, 299, 399, 599]   # Available recharge plans

    # Return True only if both conditions are satisfied
    return amount >= 50 and selected_plan in valid_plans


# Keep asking user until correct details are provided
while True:

    # Collect recharge information from the user
    recharge_amount = int(input("Enter the recharge amount: "))
    recharge_plan = int(input("Choose a plan (199, 299, 399, 599): "))

    # Validate recharge details
    if validate_recharge(recharge_amount, recharge_plan):
        print("Recharge completed successfully")
        break   # Stop the loop once recharge is valid
    else:
        print("Recharge failed")
        print("Minimum amount should be ₹50 and the plan must match available options.")
        print("Kindly re-enter the details.\n")
```

```python
# Problem Statement 2: Inventory Reorder Alert System

# The Dictionary containing the product names and their stock quantities
products = {
    "Rice": 50,
    "Wheat": 12,
    "Sugar": 8,
    "Oil": 25,
    "Salt": 30,
    "Soap": 14,
    "Shampoo": 18,
    "Toothpaste": 9,
    "Biscuits": 40,
    "Milk Powder": 6
}

# Function to check stock levels
def check_inventory(product_dict):

    # Loop through each product and its stock value
    for product, stock in product_dict.items():

        # If stock is less than 15, print reorder message
        if stock < 15:
            print(f"{product}: Reorder Alert (Stock: {stock})")

        # Otherwise, stock is sufficient
        else:
            print(f"{product}: Stock OK (Stock: {stock})")



# Call the function
check_inventory(products)
```

```python
# Problem Statement 3: Student Result Processing System

# Function to calculate average and return result
def process_result(marks_list):

    total = 0

    # Loop to calculate total marks
    for mark in marks_list:
        total += mark

    # Calculate average
    average = total / len(marks_list)

    # Print average
    print("Average Marks:", average)

    # Return Pass or Fail based on average
    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Example list of student marks
marks = [79, 47, 75, 85, 37, 90]

# Call function and store result
result = process_result(marks)

# Print final result
print("Final Result:", result)
```

```python
# Problem Statement 4: Cab Fare Estimator with Retry Option

def fare_calculator(distance, peak):
    fare = 50 + 12 * distance
    return fare * 1.25 if peak else fare


while True:
    d = float(input("Distance (km): "))
    p = input("Peak hour? (yes/no): ").lower() == "yes"

    print("Fare: ₹", fare_calculator(d, p))

    if input("Again? (yes/no): ").lower() != "yes":
        break
```

```python
# Problem Statement 5: Employee Attendance Eligibility
def attendance_eligibility(att_list):
    present = 0

    # Count number of days employee was present
    for status in att_list:
        if status == "P":
            present += 1

    # Calculate the attendance percentage
    percent = (present / len(att_list)) * 100

    # Return the eligibility based on 75% rule
    return "Eligible" if percent >= 75 else "Not Eligible"


attendance = []

# Number of the working days
n = int(input("Enter number of working days: "))

# Collect attendance for each day
for _ in range(n):
    attendance.append(input("Enter attendance (P/A): ").upper())

# Print the eligibility result
print("Employee is:", attendance_eligibility(attendance))
```

```python

# Problem Statement 6: Password Strength Checker

def password_strength(password):
    special_chars = "@#$"

    has_digit = False
    has_special = False

    # Check each character
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    # Check all conditions together
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# Test
pwd = input("Enter password: ")
print(password_strength(pwd))
```
