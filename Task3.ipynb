```python
#TASK 3
# PYTHON PROGRAMMING ASSIGNMENT DATA PROCESSING AND ANALYSIS
# PROBLEM STATEMENT1:-
# EMPLOYEE PERFORMANCE BONUS ELIGIBILITY
employees={"Ravi":92, "Anita":88,"Kiran":92,"Suresh":85}
max_score=max(employees.values())
print("Top performance Eligible for Bonus:")
for name in employees:
    if employees[name] == max_score:
        print(name,"(Score:",max_score,")")
```

```python
# PROBLEM STATEMENT 2
# SEARCHING QUERY KEYWORD ANALYSIS
query="Buy the mobile phone buy phone online"
query = query.lower()
words=query.split()
frequency={}
for word in words:
    if word in frequency:
        frequency[word]=frequency[word]+1
    else:
        frequency[word]=1
result = {}
for word in frequency:
    if frequency[word]>1:
        result[word]=frequency[word]
print(result)
```

```python
# PROBLEM STATEMENT 3
# Sensor Data Validation
sensor_readings=[3,4,7,8,10,12,5]
valid_readings=[]
for hour in range(len(sensor_readings)):
    value=sensor_readings[hour]
    if value%2==0:
       valid_readings.append((hour,value))
print("Valid Sensor Readings(Hour,Value):")
print(valid_readings)
```

```python
# PROBLEM STATEMENT 4:
# THE EMAIL DOMAIN USAGE ANALYSIS
emails=["ravi@gmail.com","anita@yahoo.com","kiran@gmail.com","suresh@gmail.com","meena@yahoo.com"]
domain_count={}
for email in emails:
    domain = email.split("@")[1]
    if domain in domain_count:
        domain_count[domain]=domain_count[domain]+1
    else:
        domain_count[domain]=1
total=len(emails)
for domain in domain_count:
    percentage=(domain_count[domain]/total)*100
    print(domain,":",str(int(percentage))+"%")
```

```python
# PROBLEM STATEMENT 5
# The Sales spike detection
sales=[1200,1500,900,2200,1400,3000]
total=0
for value in sales:
    total=total+value
average=total/len(sales)
print("Days with Sales Spikes:")
for day in range(len(sales)):
    if sales[day]>=average*1.29:
        print("Day",day+1,":",sales[day])
```

```python
# PROBLEM STATEMENT 6:
# Duplicate User ID Detection Detection
user_ids=["user1","user2","user1","user3","user1","user3"]
counts={}
for uid in user_ids:
    if uid in counts:
        counts[uid]+=1
    else:
        counts[uid]=1
for uid in counts:
    if counts[uid]>1:
        print(uid,"->",counts[uid],"times")
```
